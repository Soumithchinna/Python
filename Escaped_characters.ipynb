##### In string literals we can use escape character to associate a special meaning.

```python
#   \n for new line
da="Data\nAnalytics"
print(da)
```

```python
# \t for one tab distance
gd="Graphics\tDesign"
print(gd)
```

```python
# \b  for backspace
ai="Artificial\bintelligent"
print(ai)
```

```python
#\v vertical tab
ml="Machine\vLearning"
print(ml)
```
