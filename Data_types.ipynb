# Data Types

Data type represent the type od data present inside a variable.

In python we are not required to specify the type explicity.

Based on value provided, the type will be assigned automatically.

i.e python is dynamically typed language.


### Types of Data types(Built in) in Python

- 1. int = number
- 2. float=decimal
- 3. str = characters
- 4. complex=real + imaginary == 4+7j
- 5. bool= True/False
