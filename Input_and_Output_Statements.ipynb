# How to take multiple input at a time

```python
a,b=[int(x) for x in input("Enter 2 number:").split()]
print("Product of this:",a*b)
```

```python
a,b,c=[float(x) for x in input("Enter 3 number:").split(",")]
print("Product of this:",a*b*c)
```

# eval function

eval function will take a string and evaluate the result.

```python
x=eval("14+85+768")
x
```

```python
#input of expression
x=eval(input("Enter expression"))
x
```

```python
type(x)
```

```python
#input of list
l=eval(input("Enter list"))
print(type(l))
```

# Formated string

```python
name="Chinna"
salary=25000
friend="Sunny"
print(" hello {0} your salary is {1} and friend is {2}".format(name,salary,friend))
print(" hello {x} your salary is {y} and friend is {z}".format(x=name,y=salary,z=friend))
```
