# 6.Ternary Operator

x=first value if condition else second value.
if condition is True the first value will be considered else second value.    

```python
a,b=10,20
x=30 if a>b else 40
x
```

```python
# Taking input from keyboard and print minimum value
a=int(input("Enter the first number",))
b=int(input("Enter the second number",))
min=a if a<b else b
min
```

```python
#Taking input from keyboard and print minimum value usinf different values
a=int(input("Enter the first number",))
b=int(input("Enter the second number",))
c=int(input("Enter the third number"))
min=a if a<b and a<c else b if b<a and b<c else c 
min
```

```python
#Taking input from keyboard and print maximum value usinf different values
a=int(input("Enter the first number",))
b=int(input("Enter the second number",))
c=int(input("Enter the third number"))
max=a if a>b and a>c else b if b>a and b>c else c 
max
```
