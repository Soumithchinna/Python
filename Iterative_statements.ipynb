# Iterative statements

If we want to execute a group of statements multiple times then we should go for Iterative statements.
1.For loop
2.while loop

### 1) For loop

If we want to execute some action for every elements present in some sequence(collection of string) then we should go for for loop.
    

syntax:
for i in sequence:
    body
where sequence can be string or any collection.
body will executed on each element present in the string or collection.

```python
s="Virat Kohli"
for i in s:
    print(i, end="")
```

```python
list1=[23,"soumith",6,77,55,24]
for x in list1:
    print(x,end=",")
```

```python
#To print character present in string index wise:
name=input()
i=0
for x in name:
    print("Character present at",i,"index is",x)
    i=i+1
```

```python
print("Hello "*10)
```

```python
a=range(10)
for i in a:
    print("Hello")
```

```python
a=range(1,11)
for i in a:
    print(2*i)
```

```python
a=range(10)
for i in a:
    print(10-i)
```

```python
#print odd number 1 to 20
a=range(1,21)
for num in a:
    if num %2!=0:
        print(num)
```

```python
#take number as input and create a table
num=int(input("enter a num:"))
for i in range(1,11):
    print(num*i)
```

# 2) While loop

```python
#To print from 1 to 10 using for loop
x=1
while x<=10:
    print(x)
    x+=1
```

```python
#Display the sum of first n number:
n=int(input("Enter a num:"))
sum=0
i=1
while i<=n:
    sum=sum+i
    i+=1
print("the sum of first",n,"number is",sum)
```

# Nested loop

loop within the loop


```python
for i in range(4):
    for j in range(4):
        print("i=",i,"j=",j)
```

# Transfer statements

```python
for i in range(10):
    if i==7:
        print("Processing is enough,plz break")
        break
    print(i)    
```

```python
cart=[100,200,300,400,500,600,700,800,900]
for item in cart:
    if item>500:
        print("you get 20% discount on this order")
        break
    print(item)    
```

# Continue

we can use continue statement to skip current iteration and continue to next

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)    
```

```python
cart=[200,330,400,500,600,900,1000]
for item in cart:
    if item>=500:
        print("We cannot process this item",item)
        continue
    print(item)    
```
