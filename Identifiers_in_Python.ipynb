<!-- A name in python program is Identifier.

### Rules to define Identifier in Python

1. Only alphabets symbols are allowed(Upper and Lower case)
   A-Z
   a-z
2. 0-9 digits are allowed.
3. (_) Underscore symbol are allowed.
4.  $ is not allowed
5.  Identifier should start with character but not with numbers.
6.  Identifiers are case sensitive.
7.  we can't reserved words as identifiers.
