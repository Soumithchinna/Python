# Bytes data_types


```python
b=[10,30,40,60,78]
type(b)
```

```python
x=bytes(b)
type(x)
```

```python
for i in b:
    print(i)
```

```python
#Bytes data type is immutable 
#x[0]=80
```

```python
x[0]=5
```

# Bytesarray data Type

Its same as bytes data type but its mutable

```python
y=[20,68,40,80,50]
```

```python
ba=bytearray(y)
```

```python
for i in ba:
    print(i)
```

```python
print(ba[0])
```

```python
str(ba)
```

```python
bool(ba)
```

```python
ba[1]=76
for i in ba:
    print(i)
```

# range data type

```python
a=range(10)
a
```

```python
for i in a:
    print(i, end=" ")
```

```python
for i in range(10):
    print(i)
```

```python
c=range(2,22,2)
for i in c:
    print(i, end=" ")
```

```python
#item assigment is not possible in range.
#c[1]=20
```

```python
l=list(range(5,51,5))
print(l)
```
