# Set data structure

If we want to represent a group of unique elements then we should use set.

1.Duplicates are not allowed

2.Insertion order is not preserved

3. We can sort the elements

4.Indexing and slicing are not allowed

5.Heterogeneous objects are allowed

```python
s={10,30,40,50}
```

```python
type(s)
```

```python
list1=[10,20,20,30,40,40,50,4,4,4]
```

```python
s=set(list1)
```

```python
s
```

```python
s=set(range(5))
```

```python
s
```
